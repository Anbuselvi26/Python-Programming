**1. Write a function greetings(name,dept,message) which takes in three parameters.  Demonstrate positional arguments, default arguments, keyword arguments and special parameters using a test program.**

```python
def greetings(name, dept="CSE", message="Welcome to the department!"):
    print(f"Hello {name}, {message} ({dept})")

# Positional arguments
greetings("Anbu", "S&H", "Greetings!")

# Default arguments
greetings("Selvi")

# Keyword arguments
greetings(name="Anbuselvi", message="Hope you’re doing well!", dept="IT")

# Mixing positional and keyword arguments
greetings("Anu", message="Happy to hear from you!")
```

**2. Write a function fact(n) to determine the factorial of given n using recursion.**

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial of 5:", factorial(5))
```

**3. Write a function fib(n) to determine the nth Fibonacci number of a given n using recursion.**

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)

print("Fibonacci(6):", fib(8))
```

**4. Write a function gcd(a,b) to determine the GCD of given a and b using recursion.**

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print("GCD(78, 81):", gcd(78, 81))
```

**5. Write a function sum_upto_n(n) to determine the sum upto n using recursion.**

```python
def sum_upto_n(n):
    if n == 0:
        return 0
    return n + sum_upto_n(n - 1)

print("Sum up to 10:", sum_upto_n(12))
```

**6. Write a menu driven as earlier but this time rewrite the function calc(f,a,b) to take as input arguments a function f, operand a, operand b and applies the f on a,b as f(a,b) using lambda functions. Demonstrate using a test program for different operations.**

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Division by zero not allowed"
    else:
        return x / y

def calc(f, a, b):
    return f(a, b)

while True:
    print("\nMenu:")
    print("1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit")
    choice = int(input("Enter choice: "))

    if choice == 5:
        break

    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    if choice == 1:
        print("Result:", calc(add, a, b))
    elif choice == 2:
        print("Result:", calc(subtract, a, b))
    elif choice == 3:
        print("Result:", calc(multiply, a, b))
    elif choice == 4:
        print("Result:", calc(divide, a, b))
    else:
        print("Invalid choice!")
```

**7. Write a function conditional_apply(f1, f2, cond, x) that applies f1 to x if cond(x) is True, otherwise applies f2.**

```python
def conditional_apply(f1, f2, cond, x):
    if cond(x):
        return f1(x)
    else:
        return f2(x)

def square(num):
    return num * num

def add_ten(num):
    return num + 10

def is_even(num):
    return num % 2 == 0

print(conditional_apply(square, add_ten, is_even, 15))
print(conditional_apply(square, add_ten, is_even, 52))
```

**8. Write a function find_min_max(f,a,b) which takes as input arguments a function f, starting range a, end range b and determines the max and min of f(x) for a<=x<=b using lambda functions. Demonstrate using a test program for different functions.**

```python
def find_min_max(f, a, b):
    values = [f(x) for x in range(a, b + 1)]
    return min(values), max(values)

def quadratic(x):
    return x**2 - 4*x + 5

mn, mx = find_min_max(quadratic, -55, 55)
print(f"Min: {mn}, Max: {mx}")
```

**9. Write a function compose(f,g,x) which takes as input argument two functions f and g and returns f(g(x)). For example, $f(x)=\sqrt{x},g(x)=x^2,f(g(x))=x$. Demonstrate using a test program for different function compositions.**

```python
def compose(f, g, x):
    return f(g(x))

def square(num):
    return num * num

def double(num):
    return 2 * num

print("compose(square, double, 3):", compose(square, double, 3))
print("compose(double, square, 3):", compose(double, square, 3))
```

**10. Write a function check_fancy_number(f,n) which a takes a function f, number n and prints whether n is a Prime or Palindrome or Armstrong or Krishnamoorthy number. Demonstrate using a test program for different numbers.**

```python
def check_fancy_number(f, n):
    def is_prime(x):
        if x < 2:
            return False
        for i in range(2, int(x**0.5) + 1):
            if x % i == 0:
                return False
        return True

    def is_palindrome(x):
        return str(x) == str(x)[::-1]

    def is_armstrong(x):
        digits = [int(d) for d in str(x)]
        total = 0
        for d in digits:
            total += d ** len(digits)
        return total == x

    def is_krishnamoorthy(x):
        total = 0
        for d in str(x):
            total += f(int(d))
        return total == x

    print(f"\nChecking {n}:")
    if is_prime(n):
        print("Prime number")
    if is_palindrome(n):
        print("Palindrome number")
    if is_armstrong(n):
        print("Armstrong number")
    if is_krishnamoorthy(n):
        print("Krishnamoorthy number")

check_fancy_number(factorial, 145)
check_fancy_number(factorial, 153)
check_fancy_number(factorial, 131)
check_fancy_number(factorial, 7)
```
